Next, I want to figure out how many datapoints would be considered outliers in the dataset based on the z-score for a given point.

```python
from scipy import stats

def detect_outliers(df_without_outliers):
    #empty list to store the column names of numerical columns
    numerical_cols = []
    #iterate through the columns of the dataframe
    for col in df_without_outliers.columns:
        #if the datatype in the column is numeric then add its name to the list
        if np.issubdtype(df_without_outliers[col].dtype, np.number):
            numerical_cols.append(col)
    #z score has to be at least 3 standard deviations from mean
    threshold = 3
    #iterate through each numerical column
    for col in numerical_cols:
        #find z score of each value in the column
        z = np.abs(stats.zscore(df_without_outliers[col]))
        #get the rows with Z-scores more than the threshold
        outliers = np.where(z > threshold)
        #remove the entries with outliers from the dataframe
        df_without_outliers = df_without_outliers.drop(df_without_outliers.index[outliers])
    return df_without_outliers

# Apply the function to the dataframe
df_without_outliers = df
df_without_outliers = detect_outliers(df_without_outliers)
df_without_outliers.shape
```

For 'inm_size', 'dem_PropConEstudiosUniversitarios', and 'dem_EdadMedia' we can simply imputate the mean to handle missing values. We cannot, however, imputate the mean for categorical variables such as the 'inm_barrio' and 'inm_distrito'.

In this case, imputating the 'mode' of these categorial variables isn't reasonable because that may end up skewing the dataset.






```python
def fill_missing_values(df):
    #create a dictionary to keep track of what distritos barrios belong to
    #this iterates through every entry in the dataset to catch instances where
    #a barrio-distrito pair is correctly filled out for one entry but not another
    barrio_distrito_mapping = {}

    #populate the dictionary based on already existing key-value pairs
    for index, row in df.iterrows():
        inm_barrio, inm_distrito = row['inm_barrio'], row['inm_distrito']
        if pd.notnull(inm_barrio) and pd.notnull(inm_distrito):
            barrio_distrito_mapping[inm_barrio] = inm_distrito

    #fill in the missing distritos based on the mapping created above
    for index, row in df.iterrows():
        inm_barrio, inm_distrito = row['inm_barrio'], row['inm_distrito']
        if pd.isnull(inm_distrito) and pd.notnull(inm_barrio):
            if inm_barrio in barrio_distrito_mapping:
                df.at[index, 'inm_distrito'] = barrio_distrito_mapping[inm_barrio]

    return df


fill_missing_values(df)
```

Now, we are going to impute the most frequently seen barrio in the distrito for each null value that already has a distrito assigned to it.

```python
def fill_missing_barrios_with_mode(df):
#xfa perdona el spanglish profe

    #find the mode 'barrio' for each distrito
    mode_barrio_por_distrito = df.dropna(subset=['inm_barrio', 'inm_distrito']).groupby('inm_distrito')['inm_barrio'].agg(pd.Series.mode)

    #if a distrito has more than one mode for whatever reason just pick the first one in the list
    mode_barrio_por_distrito = mode_barrio_por_distrito.apply(lambda x: x.iloc[0] if isinstance(x, pd.Series) else x)

    #fill in missing barrio values based on the most common barrio in the distrito
    for distrito, mode_barrio in mode_barrio_por_distrito.items():
        df.loc[(df['inm_distrito'] == distrito) & (df['inm_barrio'].isnull()), 'inm_barrio'] = mode_barrio

    return df

fill_missing_barrios_with_mode(df)
```

```python
#fill in the empty spaces in those columns with the mean
df['inm_floor'] = df['inm_floor'].fillna(df['inm_floor'].mean())
df['dem_PropConEstudiosUniversitarios'] = df['dem_PropConEstudiosUniversitarios'].fillna(df['dem_PropConEstudiosUniversitarios'].mean())
df['dem_EdadMedia'] = df['dem_EdadMedia'].fillna(df['dem_EdadMedia'].mean())
df['inm_size'] = df['inm_size'].fillna(df['inm_size'].mean())
```

```python
def count_outliers(df):
    outliers_count = {}
    for column in df.columns:
        if df[column].dtype in ['int64', 'float64']:  # Only numeric columns
            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            # Counting outliers
            outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
            outliers_count[column] = len(outliers)
        else:
            outliers_count[column] = None  # Or 0 if you prefer, for non-numeric columns

    return outliers_count
count_outliers(df)
```

Now that we have an idea of how many outliers each feature has (keeping in mind that we used the 'df_count_outliers(df)' function earlier in the EDA), we can assume that a normal data distribution should not have more than 0.03% of all values on either side, given that we have chosen to calculate the outliers with three standard deviations from the mean.
